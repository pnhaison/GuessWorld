```python
import requests
import json
from IPython.display import clear_output

realWorld = requests.get("https://wordle.votee.dev:8000/random?guess=theft&seed=1234")
realWorld = "".join( [w["guess"] for w in json.loads(realWorld.text)] ).upper()
maxGuess = 5
maxLetter = 5
idx = 0
guessMatrix = ["*" * maxLetter] * maxGuess


def showMatrix(guessMatrix, realW):
    gWs=""
    for guess in guessMatrix:
        for letter, correct in zip(guess, realW):
            if letter == correct:
                letter = str(letter).upper()
            elif letter in realW:
                letter = str(letter).lower()
            elif letter not in realW and letter != "*":
                letter = "_"

            gWs = gWs + letter
        print(f"\n{gWs}")
        gWs = ""


def guessWord(guessesEDlist):
    char_ =['!','@','#','$','%','^','&','*','(',')','-','_','+','=','<','>','?','/']
    guess = input("\nGuess word: ").upper()

    if guess in guessesEDlist:
      print(f"It was guessed {guess}.")
      return guessWord(guessesEDlist)
    if len(guess) != 5:
        print(f"Your world must be {maxLetter} letters.")
        return guessWord(guessesEDlist)
    if any(letter in char_ for letter in guess):
        print(f"Invalid letter ." )
        return guessWord(guessesEDlist)
    return guess

def endGame(guesses, realW):
    clear_output()
    print(guessMatrix)

    showMatrix(guessMatrix, realW)

    if guesses == realW:
      print(f"Youw win")
    else:
      print(f"The real world : {realW}")

def main():

  for i in range(maxGuess):
      idx=i
      clear_output()
      showMatrix(guessMatrix, realWorld)

      guessMatrix[i] = guessWord(guessesEDlist=guessMatrix[:idx])
      if guessMatrix[i] == realWorld:
          break

  endGame(guessMatrix[idx], realWorld )


main()
```
